## 2. 퍼셉트론

2.1 퍼셉트론에 대해 간략하게 설명하세요.

>답: 퍼셉트론(perceptron)은 인공 신경망의 기본 단위로, 생물학적 뉴런을 모방한 단층 신경망 모델이다. 퍼셉트론은 여러 개의 입력값을 받아들이고, 각 입력값에 가중치를 곱한 후, 모든 값을 합산하여 활성화 함수(일반적으로 계단 함수)를 통해 출력을 결정한다. 이 과정에서 출력은 이진 값(0 또는 1)을 가지게 됩니다.

image.png

2.2 퍼셉트론에서 사용되는 다음 4개의 개념에 대해 설명하세요.

>입력층: 입력층은 외부에서 받은 데이터를 퍼셉트론 모델로 전달하는 역할을 한다. 입력층의 각 노드는 하나의 특징(피처)을 나타내며, 각 특징값은 퍼셉트론의 가중치와 곱해져 계산에 사용된다.

>출력층: 출력층은 퍼셉트론의 최종 결과를 출력하는 층이다. 입력층으로부터 전달된 값들이 가중치와 곱해져 합산된 후, 활성화 함수를 통해 결정된 출력값(예: 0 또는 1)이 출력층에서 반환된다.

>가중치: 가중치는 입력값에 곱해지는 값으로, 퍼셉트론이 학습을 통해 조정하는 파라미터이다. 각 가중치는 입력값이 출력에 미치는 영향을 조절하며, 학습 과정에서 오차를 최소화하기 위해 수정된다.

>편향: 편향(bias)은 가중합에 더해지는 상수 값으로, 활성화 함수의 임계값을 조절한다. 편향을 통해 퍼셉트론이 보다 유연하게 다양한 문제를 해결할 수 있도록 돕는다.

2.3 단층퍼셉트론과 다층퍼셉트론의 차이를 설명하세요.

>답: 단층퍼셉트론(single-layer perceptron)은 입력층과 출력층으로만 구성된 신경망으로, 선형적으로 구분 가능한 문제만 해결할 수 있다. 반면, 다층퍼셉트론(multilayer perceptron, MLP)은 입력층, 여러 개의 은닉층(hidden layers), 그리고 출력층으로 구성되며, 비선형 문제를 해결할 수 있는 능력을 갖추고 있다. 다층퍼셉트론은 각 층마다 활성화 함수를 사용하여 복잡한 패턴과 관계를 학습한다.

2.4 단층퍼셉트론의 한계로는 무엇이 있을 지, 다층퍼셉트론은 이를 어떻게 해결하는지에 대해 간략하게 설명하세요.

>답: 단층퍼셉트론의 주요 한계는 XOR 문제와 같은 비선형적으로 구분되는 데이터를 분류할 수 없다는 점이다. 이는 단층퍼셉트론이 선형 결정 경계만을 학습할 수 있기 때문이다. 다층퍼셉트론은 이러한 한계를 해결할 수 있다. 은닉층을 추가하고 비선형 활성화 함수를 사용함으로써, 다층퍼셉트론은 비선형 결정 경계를 학습할 수 있다.

2.5 남은 실습은 AND, OR, NAND 게이트들을 퍼셉트론으로 구현하는 것으로 마무리 짓겠습니다. 먼저 아래 코드를 완성해 AND 게이트부터 구현해보겠습니다.

출력 예시:

image.png

```python
import numpy as np

def AND(x1, x2):
    x = np.array([x1, x2]) #입력 값들을 나타내는 NumPy 배열
    w = np.array([0.5, 0.5]) #가중치 값들을 나타내는 NumPy 배열
    b = -0.7 #편향 값
    tmp = np.sum(x * w) + b
    return 0 if tmp <= 0 else 1

combinations = [[0, 0], [1, 0], [0, 1], [1, 1]]
for x1, x2 in combinations:
    print(f"{x1} AND {x2} = {AND(x1, x2)}")
```

2.6 다음으론 OR 게이트를 구현해보겠습니다.

출력 예시:

image.png

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w * x) + b
    return 0 if tmp <= 0 else 1

combinations = [[0, 0], [1, 0], [0, 1], [1, 1]]
for x1, x2 in combinations:
    print(f"{x1} OR {x2} = {OR(x1, x2)}")
```

2.7 마지막으로, NAND게이트를 구현해보겠습니다. 위에 구현한 AND 와 OR 게이트를 참고하여, NAND를 완성해주시면 됩니다.

출력 예시:

image.png

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w * x) + b
    return 0 if tmp <= 0 else 1

combinations = [[0, 0], [1, 0], [0, 1], [1, 1]]
for x1, x2 in combinations:
    print(f"{x1} NAND {x2} = {NAND(x1, x2)}")
```
