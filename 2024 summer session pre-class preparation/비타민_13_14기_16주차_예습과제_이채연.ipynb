#### 1-1. 아래 Convolution(합성곱)의 정의를 보고 합성곱의 의미를 적어주세요. (5점)

$(f*g)(t) = \int_{-\infty}^{\infty} f(\tau)g(t-\tau)\, d\tau$

답) 합성곱(Convolution)은 두 함수 \(f(x)\)와 \(g(x)\)가 있을 때, 하나의 함수를 뒤집고(좌우 반전) 다른 함수에 대해서 이동하면서 두 함수의 곱을 적분하여 새로운 함수를 생성하는 연산입니다.
- 합성곱 연산은 주어진 신호를 다른 신호(커널 함수)로 변형하거나 필터링하는 과정으로 이해할 수 있습니다.
- 예를 들어, 이미지 처리에서 합성곱은 이미지의 특정 패턴(엣지, 블러 등)을 강조하거나 제거하는 데 사용됩니다.
- 시간 도메인에서는 신호의 특성을 다른 신호와 결합하여 특정 특징을 추출하거나, 신호를 평활화(스무딩)하거나 변형시키는 데 사용됩니다.

쉽게 말해, 합성곱은 하나의 함수가 다른 함수와 상호작용하여 새로운 출력을 생성하는 방법을 제공하는 연산입니다. 이는 주로 신호 처리, 이미지 처리, 그리고 심층 신경망(특히 CNN, Convolutional Neural Networks)에서 중요한 역할을 합니다.

#### 1-2. CNN의 주요 용어인 Filter, Channel, Stride, Feature map에 대해 각각 간단하게 한 줄로 설명해주세요. (5점)

답)
- **Filter**: 입력 데이터에 적용되어 특징을 추출하는 가중치 행렬(커널).
- **Channel**: 이미지나 입력 데이터의 깊이를 나타내며, 예를 들어 RGB 이미지는 3개의 채널(R, G, B)로 구성됨.

- **Feature map**: 필터가 입력 데이터와 합성곱 연산을 수행한 결과로 생성된 출력 데이터, 특징이 추출된 맵.

#### 1-7. 아래 reference를 참조하여 각 파라미터가 무엇을 의미하는지 작성하세요. (5점)

- `torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride=1, padding=0)`
- Reference : https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html


답)
- in_channels : 입력 데이터의 채널 수를 지정합니다. 예를 들어, RGB 이미지의 경우 3(레드, 그린, 블루 채널)입니다.
- out_channels : 합성곱 연산의 결과로 생성될 출력 채널의 수를 지정합니다. 이는 필터의 수와 관련이 있으며, 각 필터가 출력 채널을 하나 생성합니다.
- kernel_size : 합성곱 연산에 사용될 커널(필터)의 크기를 지정합니다. 보통 정수나 (높이, 너비) 튜플로 주어집니다. 예를 들어, kernel_size=3은 3x3 크기의 커널을 의미합니다.
- stride : 필터가 입력 데이터에서 이동하는 간격을 지정합니다. 스트라이드가 2면 필터는 두 칸씩 이동하여 출력 맵을 생성합니다.
- padding : 입력 데이터의 가장자리에 추가할 패딩의 크기를 지정합니다. 패딩을 추가하면 출력 맵의 크기가 더 커질 수 있습니다. 예를 들어, padding=1은 입력의 모든 가장자리에 1픽셀의 패딩을 추가합니다.

#### 1-8. 아래 코드의 빈칸을 채워주세요. (3점)

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

####################### 문제 #######################

# 크기가 20 * 1 * 28 * 28인 임의의 텐서 생성
input = torch.randn(20, 1, 28, 28)  # 배치 크기 20, 채널 1, 높이 28, 너비 28

# 첫번째 conv layer 구성 : in_channel=1, out_channel=3, kernel_size=3, stride=1
conv1 = nn.Conv2d(in_channels=1, out_channels=3, kernel_size=3, stride=1)
print(conv1) # 구성한 conv layer 확인
output = conv1(input)

# 첫번째 conv layer을 통과한 tensor의 shape 확인
print(output.shape)  # 예상되는 출력: torch.Size([20, 3, 26, 26])

# 두번째 conv layer 구성 : in_channel=3, out_channel=9, kernel_size=5, stride=1
conv2 = nn.Conv2d(in_channels=3, out_channels=9, kernel_size=5, stride=1)
print(conv2) # 구성한 conv layer 확인
output_2 = conv2(output)

# 두번째 conv layer을 통과한 tensor의 shape 확인
print(output_2.shape)  # 예상되는 출력: torch.Size([20, 9, 22, 22])
```

#### 1-9. 아래 코드의 빈칸을 채워주세요.  (6점)

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 3, 3, 1)  # 입력 채널 1, 출력 채널 3, 커널 크기 3, 스트라이드 1
        self.pool1 = nn.MaxPool2d(2, 2)     # 첫 번째 MaxPooling 추가
        self.conv2 = nn.Conv2d(3, 9, 5, 1)  # 입력 채널 3, 출력 채널 9, 커널 크기 5, 스트라이드 1
        self.pool2 = nn.MaxPool2d(2, 2)     # 두 번째 MaxPooling 추가
        self.fc = nn.Linear(9 * 4 * 4, 10) # Linear layer input size 수정

    def forward(self, x):
        print("연산 전", x.size())
        out = self.conv1(x)               # 첫 번째 Conv2d 레이어
        out = F.relu(out)
        out = self.pool1(out)             # 첫 번째 MaxPooling 적용
        print("conv1 연산 후", out.size())

        out = self.conv2(out)             # 두 번째 Conv2d 레이어
        out = F.relu(out)
        out = self.pool2(out)             # 두 번째 MaxPooling 적용
        print("conv2 연산 후", out.size())

        out = out.view(out.size(0), -1)   # Flatten
        print("차원 감소 후", out.size())

        out = self.fc(out)                # Fully Connected layer
        print("fc 연산 후", out.size())

        return out

# 입력 텐서 생성
input = torch.randn(20, 1, 28, 28)  # 배치 크기 20, 채널 1, 높이 28, 너비 28

# 모델 생성 및 출력 확인
cnn = CNN()
output = cnn(input)
print('output shape: {}'.format(output.shape))
```

#### 문제 2. 다음 그림은 LeNet의 구조입니다. LeNet의 C3 layer가 S2의 모든 feature map과 연결되지 않는 이유는 무엇일까요? (3점)


답:

LeNet의 C3 레이어가 S2 레이어의 모든 feature map과 연결되지 않는 이유는 **부분 연결(partial connectivity) 구조** 때문입니다. 이 구조는 **연산의 효율성**과 **모델의 표현력**을 높이기 위한 전략입니다.

### 자세한 설명:
부분 연결 구조:

- LeNet-5의 C3 레이어는 S2 레이어의 모든 feature map과 완전히 연결된 것이 아니라, 일부 feature map만 연결됩니다.
- 이 부분 연결 구조는 모델의 파라미터 수를 줄이고 계산 복잡성을 낮추기 위해 설계되었습니다.

이유:

1.  연산 효율성: 전체 연결 구조를 사용하면 필터의 수와 파라미터 수가 기하급수적으로 증가할 수 있습니다. 부분 연결은 이러한 문제를 해결하고 계산을 보다 효율적으로 만들 수 있습니다.
2.  표현력 향상: 특정 feature map들 간의 연결을 통해 중요한 특징을 강조할 수 있습니다. 부분 연결은 특정 지역의 패턴을 추출하는 데 집중할 수 있도록 해줍니다.
3.  학습 안정성: 전체 연결이 아니라 부분 연결을 사용하면 과적합(overfitting) 위험을 줄이고, 학습 과정에서 보다 안정적인 결과를 얻을 수 있습니다.


LeNet-5의 C3 레이어 예시: C3 레이어는 S2 레이어의 6개의 feature map 중 일부와 연결됩니다. 예를 들어, C3 레이어의 각 feature map은 S2 레이어의 특정 feature map들과만 연결될 수 있습니다. 이러한 연결 패턴은 실험적으로 설정되며, 모델의 성능과 효율성을 최적화하는 데 도움이 됩니다.





이러한 부분 연결 구조는 CNN의 설계에서 중요한 개념으로, 네트워크의 구조와 학습의 효율성을 조절하는 데 기여합니다.

#### 문제 7. AlexNet의 첫 번째 convolution layer는 227x227x3 크기의 입력 이미지를 받으며, 이 layer는 11x11 크기의 필터 96개를 사용하고, stride는 4, padding은 0입니다. 첫 번째 convolution layer의 출력 feature map의 크기를 구해주세요. (7점)

답:
$\[
\text{Output Size} = \frac{n - f + 2p}{s} + 1
\]$

$\text{여기서:}$
\begin{align*}
n & = \text{입력 크기} \\
f & = \text{필터 크기} \\
p & = \text{패딩} \\
s & = \text{스트라이드}
\end{align*}

$\text{위 값을 대입하면:}$
\begin{align*}
\text{Output Height} &= \frac{227 - 11 + 2 \cdot 0}{4} + 1 \\
&= \frac{216}{4} + 1 \\
&= 54 + 1 \\
&= 55
\end{align*}

$\text{따라서 출력 feature map의 크기는 } 55 \times 55 \text{ 입니다.}$
