### 장점
1. **빠른 수렴 속도**: SGD는 매 반복마다 하나의 샘플을 사용하여 기울기를 계산하기 때문에, 대용량 데이터셋에 대해서도 빠르게 학습할 수 있습니다. 이는 배치 크기가 큰 경우에 비해 더 빠른 초기 수렴을 가능하게 합니다.

2. **메모리 효율성**: 매번 전체 데이터셋을 사용하지 않고 하나의 샘플만을 사용하기 때문에, 메모리 사용량이 적습니다. 이는 특히 대규모 데이터셋을 처리할 때 유용합니다.

3. **국소 최적점 탈출 가능성**: SGD의 특유의 노이즈(각 반복마다 다른 샘플을 사용하기 때문에 발생)는 국소 최적점(local minima)을 탈출하고 더 나은 전역 최적점(global minimum)에 도달할 가능성을 높여줍니다.

### 단점
1. **수렴의 불안정성**: SGD는 매 반복마다 계산된 기울기가 달라지기 때문에, 목적 함수의 값이 진동하거나 변동할 수 있습니다. 이는 학습 과정에서 수렴의 불안정성을 야기할 수 있습니다.

3. **정확한 기울기 정보 부족**: 전체 데이터셋에 기반한 기울기 정보를 사용하지 않기 때문에, SGD는 기울기의 방향을 정확하게 나타내지 못할 수 있습니다. 이는 최적화의 효율성을 저하시킬 수 있습니다.

결과적으로, 모멘텀은 경사 하강법의 수렴 속도를 증가시키고, 진동을 줄이며, 국소 최적점에서 벗어나 전역 최적점으로 더 잘 도달하게 하는 효과가 있습니다.

**1-8.**  다음은 Adam 옵티마이저의 코드입니다. 빈칸을 채워 올바른 코드를 완성하세요. (5점)

```python
def adam_update(params, grads, m, v, t, learning_rate=0.01, beta1=0.9, beta2=0.999, epsilon=1e-8):
    for i in range(len(params)):
        # 모멘텀 추정값 업데이트
        m[i] = beta1 * m[i] + (1 - beta1) * grads[i]

        # 제곱 기울기 추정값 업데이트
        v[i] = beta2 * v[i] + (1 - beta2) * (grads[i] ** 2)

        # 보정된 모멘텀 추정값
        m_hat = m[i] / (1 - beta1 ** t)

        # 보정된 제곱 기울기 추정값
        v_hat = v[i] / (1 - beta2 ** t)

        # 파라미터 업데이트
        params[i] -= learning_rate * m_hat / (v_hat ** 0.5 + epsilon)

    return params, m, v

params = [1.0, 2.0]
grads = [0.1, -0.2]
m = [0.0, 0.0]
v = [0.0, 0.0]
t = 1
updated_params, updated_m, updated_v = adam_update(params, grads, m, v, t)
print(updated_params)  # 예상 출력: [0.999, 2.001]
```

**2-6.** 해당 코드는 각 층의 활성화값들의 분포가 어떻게 변화하는지  
관찰하는 코드입니다. 주석에 담긴 정보를 활용하여 빈 칸을 채워주세요. (3점)

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def ReLU(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

input_data = np.random.randn(1000, 100)  # 1000개의 데이터, 표준정규분포로 1000행 100열 행렬 생성
node_num = 100  # 각 은닉층의 노드(뉴런) 수
hidden_layer_size = 5  # 은닉층이 5개
activations = {}  # 이곳에 활성화 결과를 저장

x = input_data

for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i-1]

    # Xavier 초깃값 - 표준편차 1/√n
    w = np.random.randn(node_num, node_num) * (1 / np.sqrt(node_num))

    a = np.dot(x, w)

    # 활성화 함수 - ReLU
    z = ReLU(a)

    activations[i] = z

# 히스토그램 그리기
for i, a in activations.items():
    plt.subplot(1, len(activations), i+1)
    plt.title(str(i+1) + "-layer")
    if i != 0: plt.yticks([], [])
    plt.xlim(0.0, 1.0)
    plt.ylim(0, 7000)
    plt.hist(a.flatten(), 30, range=(0,1))
plt.show()
```

**1-6.** 다음 코드는 L1 정규화를 기하적으로 표현합니다. 빈 코드를 완성해주세요. (5점)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the L1 norm function
def l1_norm(w):
    return np.sum(np.abs(w), axis=0)

# Create a grid of points
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)

# Calculate the L1 norm for each point in the grid
Z = l1_norm(np.array([X, Y]))

# Plot the L1 norm contours
plt.figure(figsize=(8, 8))
contours = plt.contour(X, Y, Z, levels=15, cmap='viridis')
plt.clabel(contours, inline=True, fontsize=8)
plt.title('L1 Regularization Contour Plot')
plt.xlabel('w1')
plt.ylabel('w2')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.show()
```

**1-7.** 다음은 tanh의 시각화 코드입니다. 출력 이미지를 참고하여 입력 데이터가 0에서 멀리에 위치해 있을때 우려되는 점을 간단히 서술해주세요. (2점)

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

x = np.linspace(-5, 5, 400)
y = tanh(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='tanh(x)', color='blue')
plt.title('Tanh Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

(답 입력) tanh 함수는 입력 데이터가 0에서 멀리 떨어져 있을 때 출력이 -1 또는 1에 가까워지는 포화 현상을 보입니다. 이 경우, 입력 데이터의 변화가 출력에 거의 영향을 주지 않아, 기울기 소실(vanishing gradient) 문제가 발생할 수 있습니다. 이는 신경망 학습 과정에서 기울기가 거의 0에 가까워져서 가중치 업데이트가 잘 이루어지지 않게 되어, 학습 속도가 느려지거나 멈출 수 있습니다.

```python
# 실행할 필요 없습니다.
import numpy as np
from skopt import gp_minimize
from skopt.space import Real
import matplotlib.pyplot as plt

# 베이지안 최적화를 적용할 함수
def objective(x):
    return (x[0] - 2) ** 2  # x is a list, so we take x[0]

search_space = [Real(-5.0, 5.0, name='x')]

# 베이지안 최적화
result = gp_minimize(objective, search_space, n_calls=20, random_state=42)

# 결과
print(f"Best parameter: {result.x}")
print(f"Best objective value: {result.fun}")

# 시각화
plt.figure(figsize=(10, 6))
plt.plot(result.func_vals, marker='o', color='b')
plt.title('Bayesian Optimization Process')
plt.xlabel('Iteration')
plt.ylabel('Objective Value')
plt.grid(True)
plt.show()
```

이 코드는 베이지안 최적화를 사용하여 단순한 2차 함수의 최소값을 찾는 과정과 그 결과를 시각화한 결과입니다.

이 목적함수는 x가 2일 때 최소값을 가집니다.
- f(x) = (x-2)^2

결과를 참고하여 베이지안 최적화의 결과가 어떤거 같은지 2줄 정도로 서술해주세요. (4점)
