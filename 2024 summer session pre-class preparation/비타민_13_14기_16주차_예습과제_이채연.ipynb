- padding : ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Ïùò Í∞ÄÏû•ÏûêÎ¶¨Ïóê Ï∂îÍ∞ÄÌï† Ìå®Îî©Ïùò ÌÅ¨Í∏∞Î•º ÏßÄÏ†ïÌï©ÎãàÎã§. Ìå®Îî©ÏùÑ Ï∂îÍ∞ÄÌïòÎ©¥ Ï∂úÎ†• ÎßµÏùò ÌÅ¨Í∏∞Í∞Ä Îçî Ïª§Ïßà Ïàò ÏûàÏäµÎãàÎã§. ÏòàÎ•º Îì§Ïñ¥, padding=1ÏùÄ ÏûÖÎ†•Ïùò Î™®Îì† Í∞ÄÏû•ÏûêÎ¶¨Ïóê 1ÌîΩÏÖÄÏùò Ìå®Îî©ÏùÑ Ï∂îÍ∞ÄÌï©ÎãàÎã§.

#### 1-8. ÏïÑÎûò ÏΩîÎìúÏùò ÎπàÏπ∏ÏùÑ Ï±ÑÏõåÏ£ºÏÑ∏Ïöî. (3Ï†ê)

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

####################### Î¨∏Ï†ú #######################

# ÌÅ¨Í∏∞Í∞Ä 20 * 1 * 28 * 28Ïù∏ ÏûÑÏùòÏùò ÌÖêÏÑú ÏÉùÏÑ±
input = torch.randn(20, 1, 28, 28)  # Î∞∞Ïπò ÌÅ¨Í∏∞ 20, Ï±ÑÎÑê 1, ÎÜíÏù¥ 28, ÎÑàÎπÑ 28

# Ï≤´Î≤àÏß∏ conv layer Íµ¨ÏÑ± : in_channel=1, out_channel=3, kernel_size=3, stride=1
conv1 = nn.Conv2d(in_channels=1, out_channels=3, kernel_size=3, stride=1)
print(conv1) # Íµ¨ÏÑ±Ìïú conv layer ÌôïÏù∏
output = conv1(input)

# Ï≤´Î≤àÏß∏ conv layerÏùÑ ÌÜµÍ≥ºÌïú tensorÏùò shape ÌôïÏù∏
print(output.shape)  # ÏòàÏÉÅÎêòÎäî Ï∂úÎ†•: torch.Size([20, 3, 26, 26])

# ÎëêÎ≤àÏß∏ conv layer Íµ¨ÏÑ± : in_channel=3, out_channel=9, kernel_size=5, stride=1
conv2 = nn.Conv2d(in_channels=3, out_channels=9, kernel_size=5, stride=1)
print(conv2) # Íµ¨ÏÑ±Ìïú conv layer ÌôïÏù∏
output_2 = conv2(output)

# ÎëêÎ≤àÏß∏ conv layerÏùÑ ÌÜµÍ≥ºÌïú tensorÏùò shape ÌôïÏù∏
print(output_2.shape)  # ÏòàÏÉÅÎêòÎäî Ï∂úÎ†•: torch.Size([20, 9, 22, 22])
```

#### 1-9. ÏïÑÎûò ÏΩîÎìúÏùò ÎπàÏπ∏ÏùÑ Ï±ÑÏõåÏ£ºÏÑ∏Ïöî.  (6Ï†ê)

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 3, 3, 1)  # ÏûÖÎ†• Ï±ÑÎÑê 1, Ï∂úÎ†• Ï±ÑÎÑê 3, Ïª§ÎÑê ÌÅ¨Í∏∞ 3, Ïä§Ìä∏ÎùºÏù¥Îìú 1
        self.pool1 = nn.MaxPool2d(2, 2)     # Ï≤´ Î≤àÏß∏ MaxPooling Ï∂îÍ∞Ä
        self.conv2 = nn.Conv2d(3, 9, 5, 1)  # ÏûÖÎ†• Ï±ÑÎÑê 3, Ï∂úÎ†• Ï±ÑÎÑê 9, Ïª§ÎÑê ÌÅ¨Í∏∞ 5, Ïä§Ìä∏ÎùºÏù¥Îìú 1
        self.pool2 = nn.MaxPool2d(2, 2)     # Îëê Î≤àÏß∏ MaxPooling Ï∂îÍ∞Ä
        self.fc = nn.Linear(9 * 4 * 4, 10) # Linear layer input size ÏàòÏ†ï

    def forward(self, x):
        print("Ïó∞ÏÇ∞ Ï†Ñ", x.size())
        out = self.conv1(x)               # Ï≤´ Î≤àÏß∏ Conv2d Î†àÏù¥Ïñ¥
        out = F.relu(out)
        out = self.pool1(out)             # Ï≤´ Î≤àÏß∏ MaxPooling Ï†ÅÏö©
        print("conv1 Ïó∞ÏÇ∞ ÌõÑ", out.size())

        out = self.conv2(out)             # Îëê Î≤àÏß∏ Conv2d Î†àÏù¥Ïñ¥
        out = F.relu(out)
        out = self.pool2(out)             # Îëê Î≤àÏß∏ MaxPooling Ï†ÅÏö©
        print("conv2 Ïó∞ÏÇ∞ ÌõÑ", out.size())

        out = out.view(out.size(0), -1)   # Flatten
        print("Ï∞®Ïõê Í∞êÏÜå ÌõÑ", out.size())

        out = self.fc(out)                # Fully Connected layer
        print("fc Ïó∞ÏÇ∞ ÌõÑ", out.size())

        return out

# ÏûÖÎ†• ÌÖêÏÑú ÏÉùÏÑ±
input = torch.randn(20, 1, 28, 28)  # Î∞∞Ïπò ÌÅ¨Í∏∞ 20, Ï±ÑÎÑê 1, ÎÜíÏù¥ 28, ÎÑàÎπÑ 28

# Î™®Îç∏ ÏÉùÏÑ± Î∞è Ï∂úÎ†• ÌôïÏù∏
cnn = CNN()
output = cnn(input)
print('output shape: {}'.format(output.shape))
```
